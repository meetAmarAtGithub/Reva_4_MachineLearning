```python
titanic = sns.load_dataset('titanic')
titanic.tail(5)
```

```python
from IPython.core.interactiveshell  import InteractiveShell
InteractiveShell.ast_node_interactivity='all'
```

```python
titanic.drop(['alive','class','embark_town','alone','adult_male','who'], axis=1, inplace=True)
```

```python
titanic.head(5)
```

```python
titanic.info()
```

```python
titanic.isna().sum()
```

```python
titanic_age_missing = titanic[titanic['age'].isna()]
titanic_age_missing.head()
```

```python
titanic['age'] = titanic['age'].fillna(titanic['age'].median())
titanic.info()
```

```python
titanic_embark_missing = titanic[titanic['embarked'].isna()]
titanic_embark_missing
```

```python
titanic['embarked']=titanic['embarked'].fillna(titanic['embarked'].mode()[0])
```

```python
titanic.deck = titanic.deck.astype('object')
```

```python
titanic['deck']= titanic['deck'].fillna('missing')
```

```python
titanic = titanic.rename(columns= {"sex":"gender", "sibsp":'siblings', "parch":"parents_child", "Survived":'survived'})
titanic.head(5)
```

```python
titanic.info
```

```python
#Dummy coding for categorical variables
for x in titanic.columns:
    if titanic[x].dtype == "object":
        titanic[x] =pd.Categorical(titanic[x]).codes
```

```python
titanic.tail(5)
```

```python
#Import SK Learn model for doing ML Modeling
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn import tree
```

```python
x = titanic.drop(['survived'],axis=1, inplace=False)
y= titanic['survived']
x.shape
print('\n')
y.shape
```

```python
#Create training and test data
train_x,test_x,train_y,test_y= train_test_split(x,y,test_size=0.2,random_state=999)
train_x.shape
train_y.shape
test_x.shape
test_y.shape
```

```python
#Create instant of the model
myTree = tree.DecisionTreeClassifier(min_samples_leaf=50, min_samples_split=200)

#fit the model
myTree.fit(train_x,train_y)

# Predict using the trained model
predicted = myTree.predict(test_x)
predicted
```

```python
myTree.predict_proba(test_x)
```

```python
#Evaluate model perfomance
from sklearn import metrics
print(metrics.classification_report(test_y, predicted))
```

# Using Class weight

```python
from sklearn.utils import class_weight
#Using Class weight

#Create instant of the model
myTree = tree.DecisionTreeClassifier(min_samples_leaf=50, min_samples_split=200, class_weight='balanced')

#fit the model
myTree.fit(train_x,train_y)

# Predict using the trained model
predicted = myTree.predict(test_x)
predicted
```

```python
myTree.predict_proba(test_x)
```

```python
#Evaluate model perfomance
from sklearn import metrics
print(metrics.classification_report(test_y, predicted))
```

```python
#Confusion matrix plot using seaborn
df_confusion = metrics.confusion_matrix(test_y, predicted)
df_confusion
```

```python
sns.heatmap(df_confusion, cmap='Blues', xticklabels=['prediction no','prediction yes'], yticklabels=['Actual No','Actual Yes'],annot=True, fmt='d')
plt.show()
```

```python
metrics.accuracy_score(test_y, predicted)
```

```python
#Create training and test data
dot_data = tree.export_graphviz(myTree, out_file=None, feature_names=x.columns, class_names=['0','1'], filled=True, rounded=True, special_characters=True)


graph = graphviz.Source(dot_data)
graph
```

# Threshold Optimization



```python
predicted_prob = myTree.predict_proba(test_x)
testcase = np.round(predicted_prob,3)
testcase
```

```python
testcase_prediction = testcase[1] >= 0.15
testcase_prediction
```

```python
new_y_test
```

```python
#Evaluate model perfomance
from sklearn import metrics
print(metrics.classification_report(test_y, new_y_test))
```

# Hyper parameter optimization

```python
myTree = tree.DecisionTreeClassifier(random_state=99, class_weight='balanced')
myTree
```

```python
my_max_depth = [2,3,4,5,10]
my_criterion = ['gini','entropy']
my_min_sample_leaf = [2,5,10,15,20,25]
my_min_sample_split = [2,5,10,15,50,100]
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(estimator=myTree, cv=5, scoring='recall', 
                    param_grid= dict(max_depth = my_max_depth, criterion= my_criterion ,
                                     min_samples_leaf = my_min_sample_leaf, min_samples_split = my_min_sample_split))
```

```python
grid.fit(train_x, train_y)
```

```python
grid.best_params_
```

```python
grid.cv_results_
```

```python
resulsts = pd.DataFrame(grid.cv_results_)
resulsts.to_csv("resulsts.csv")
```
